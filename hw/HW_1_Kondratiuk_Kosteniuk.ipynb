```python
import pandas as pd

diabetes_data = pd.read_csv("../data/diabetes.csv")
diabetes_data.head(2)
```

```python
# Схоже що тільки Outcome колонка - категорійна
diabetes_data_numbers = diabetes_data.drop(columns=["Outcome"])
diabetes_data_categorical = diabetes_data[["Outcome"]].copy()
```

 Підрахувати основні статистичні характеристики для кожної з чисельних
 ознак:

```python
# Мода, медіана, математичне очікування;
stats_df = pd.DataFrame(
    {
        "mean": diabetes_data_numbers.mean(),
        "median": diabetes_data_numbers.median(),
        "mode": diabetes_data_numbers.mode().iloc[0],
    }
).T
stats_df.head(3)
```

```python
# Дисперсія;
var_df = diabetes_data_numbers.var().to_frame(name="variance")
var_df.head(10)

# Insulin має дуже велику дисперсію відносно інших ознак.
```

```python
# 25%, 75%, 50% квантилі.
quantiles_df = diabetes_data_numbers.quantile([0.25, 0.5, 0.75]).T
quantiles_df.head(10)
```
